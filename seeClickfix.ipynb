```python
import requests
import time

base_url = "https://seeclickfix.com/api/v2/issues"
all_results = []
page = 1
per_page = 100  # max allowed

bbox_params = {
    "min_lat": 42.64756504177151,
    "min_lng": -73.79911422729491,
    "max_lat": 42.669594022033735,
    "max_lng": -73.75173568725592,
    "after": "2025-01-01T00:00:00Z",
    "before": "2025-12-31T23:59:59Z",
    "status": "all"
}

while True:
    params = {
        **bbox_params,
        "page": page,
        "per_page": per_page
    }
    response = requests.get(base_url, params=params)
    if response.status_code != 200:
        raise Exception(f"Error: {response.status_code}, {response.text}")
    data = response.json()
    issues = data.get("issues", [])
    if not issues:
        break
    all_results.extend(issues)
    print(f"Fetched page {page}, total issues so far: {len(all_results)}")
    page += 1
    time.sleep(0.2)  # small delay to be nice to API

# Deduplicate by issue ID
unique_issues = {issue['id']: issue for issue in all_results}
print(f"Total unique issues fetched: {len(unique_issues)}")
```

```python
# unique_issues is a dict keyed by issue ID
cleaned = [
    {
        "id": issue.get("id"),
        "status": issue.get("status"),
        "summary": issue.get("summary"),
        "description": issue.get("description"),
        "rating": issue.get("rating"),
        "lat": issue.get("lat"),
        "lng": issue.get("lng"),
        "address": issue.get("address"),
        "created_at": issue.get("created_at"),
        "acknowledged_at": issue.get("acknowledged_at"),
        "closed_at": issue.get("closed_at"),
        "reopened_at": issue.get("reopened_at"),
        "url": issue.get("url"),
        "request_type_title": issue.get("request_type", {}).get("title")  # nested title
    }
    for issue in unique_issues.values()
]
```
